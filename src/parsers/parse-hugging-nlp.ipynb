```python
import re
import os
```

```python
def clean_html_and_links(file_path, output_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        content = file.read()
    emoji_pattern = re.compile(
        "["
        u"\U0001F600-\U0001F64F"  # Emoticons
        u"\U0001F300-\U0001F5FF"  # Misc Symbols and Pictographs
        u"\U0001F680-\U0001F6FF"  # Transport & Map Symbols
        u"\U0001F700-\U0001F77F"  # Alchemical Symbols
        u"\U0001F780-\U0001F7FF"  # Geometric Shapes Extended
        u"\U0001F800-\U0001F8FF"  # Supplemental Arrows-C
        u"\U0001F900-\U0001F9FF"  # Supplemental Symbols & Pictographs
        u"\U0001FA00-\U0001FA6F"  # Chess Symbols
        u"\U0001FA70-\U0001FAFF"  # Symbols & Pictographs Extended-A
        u"\U00002700-\U000027BF"  # Dingbats
        u"\U000024C2-\U0001F251"  # Enclosed Characters
        "]+",
        flags=re.UNICODE
    )
    
    content = re.sub(r'<div.*?>.*?</div>', '', content, flags=re.DOTALL)
    content = re.sub(r'\[([^\]]+)\]\s*\([^\)]+\)', r'\1', content)
    content = re.sub(r'<[a-zA-Z0-9]+[^>]*>.*?</[a-zA-Z0-9]+>', '', content, flags=re.DOTALL)
    content = re.sub(r'<[a-zA-Z0-9]+[^>]*/>', '', content)
    content = re.sub(r'<[^>]*>', '', content)
    content = emoji_pattern.sub(r'', content)

    with open(output_path, 'w', encoding='utf-8') as file:
        file.write(content)

# input_file = '1.mdx'  # Input file with HTML and links
# output_file = 'cleaned_example.txt'  # Output cleaned file
# clean_html_and_links(input_file, output_file)
```

```python
source_folder = './hugging_face/course/chapters/ru/'
target_folder = './hugging_face/nlp-course/'

for root, folders, files in os.walk(source_folder): 
    if len(files) > 0:
        chapter = root.split('/')[-1]
        chapter_folder = os.path.join(target_folder, chapter)
        !mkdir $chapter_folder
        for file in files:
            if file.endswith('.mdx'):
                output_file = f"{file.split('.')[0]}.txt"
                clean_html_and_links(os.path.join(root, file), 
                                     os.path.join(chapter_folder, output_file))
                print(f"Saved to {os.path.join(chapter_folder, output_file)}")
```
