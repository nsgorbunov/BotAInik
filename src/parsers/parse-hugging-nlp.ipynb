```python
import os
import re
```

```python
def clean_html_and_links(file_path, output_path):
    with open(file_path, "r", encoding="utf-8") as file:
        content = file.read()
    emoji_pattern = re.compile(
        "["
        "\U0001f600-\U0001f64f"  # Emoticons
        "\U0001f300-\U0001f5ff"  # Misc Symbols and Pictographs
        "\U0001f680-\U0001f6ff"  # Transport & Map Symbols
        "\U0001f700-\U0001f77f"  # Alchemical Symbols
        "\U0001f780-\U0001f7ff"  # Geometric Shapes Extended
        "\U0001f800-\U0001f8ff"  # Supplemental Arrows-C
        "\U0001f900-\U0001f9ff"  # Supplemental Symbols & Pictographs
        "\U0001fa00-\U0001fa6f"  # Chess Symbols
        "\U0001fa70-\U0001faff"  # Symbols & Pictographs Extended-A
        "\U00002700-\U000027bf"  # Dingbats
        "\U000024c2-\U0001f251"  # Enclosed Characters
        "]+",
        flags=re.UNICODE,
    )

    content = re.sub(r"<div.*?>.*?</div>", "", content, flags=re.DOTALL)
    content = re.sub(r"\[([^\]]+)\]\s*\([^\)]+\)", r"\1", content)
    content = re.sub(
        r"<[a-zA-Z0-9]+[^>]*>.*?</[a-zA-Z0-9]+>", "", content, flags=re.DOTALL
    )
    content = re.sub(r"<[a-zA-Z0-9]+[^>]*/>", "", content)
    content = re.sub(r"<[^>]*>", "", content)
    content = emoji_pattern.sub(r"", content)

    with open(output_path, "w", encoding="utf-8") as file:
        file.write(content)


# input_file = '1.mdx'  # Input file with HTML and links
# output_file = 'cleaned_example.txt'  # Output cleaned file
# clean_html_and_links(input_file, output_file)
```

```python
source_folder = "./hugging_face/course/chapters/ru/"
target_folder = "./hugging_face/nlp-course/"

for root, folders, files in os.walk(source_folder):
    if len(files) > 0:
        chapter = root.split("/")[-1]
        chapter_folder = os.path.join(target_folder, chapter)
        !mkdir $chapter_folder
        for file in files:
            if file.endswith(".mdx"):
                output_file = f"{file.split('.')[0]}.txt"
                clean_html_and_links(
                    os.path.join(root, file), os.path.join(chapter_folder, output_file)
                )
                print(f"Saved to {os.path.join(chapter_folder, output_file)}")
```
